```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
np.random.seed(0)
```

```python
m = 100
X = np.linspace(0,10,m).reshape(m,1)
y = X+np.random.randn(m,1)
plt.scatter(X,y)
```

```python
reg = linear_model.LinearRegression()
```

```python
reg.fit(X,y)
```

```python
reg.score(X,y)
```

```python
new_y = reg.predict(X)
```

```python
plt.scatter(X,y)
plt.plot(X,new_y,c="r")
```
